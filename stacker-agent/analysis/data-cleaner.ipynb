$$\sigma(z) = \frac{1}{1+e^{-z}} = \frac{1}{1+\frac{1}{e^z}} = \frac{1}{\frac{e^z}{e^z}+\frac{1}{e^z}} = \frac{1}{\frac{e^z+1}{e^z}} = \frac{e^z}{1+e^z} = \frac{e^z}{e^0+e^z}$$

If $z = W_kX+b$ where $W_k$ are the weights $W$ associated with a class $k$, then rewrite $\frac{e^z}{e^0+e^z}$ as:

$$\frac{e^{W_kX+b}}{e^{W_0X+b}+e^{W_1X+b}}$$

which is the same as

$$\frac{e^{W_kX+b}}{\sum_{n=0}^1 e^{W_nX+b}}$$

so therefore if there are only 2 classes $\frac{1}{1+e^{-z}} = \frac{e^z_i}{\sum_{n=0}^K e^z_n}, \text{for } i = 1,...,K$
